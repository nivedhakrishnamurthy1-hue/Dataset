```python
print("\nK.Nivedha(24BAD082)")
print("\nEXPERIMENT 1")
print("\nSCENARIO 4: CUSTOMER PERSONALITY\n")

import pandas as pd
import matplotlib.pyplot as plt

#Load dataset correctly
df = pd.read_csv(r"C:\Users\nived\Downloads\marketing_campaign.csv", sep="\t")

#Inspect dataset
print("FIRST 5 ROWS:")
print(df.head())

print("\nLAST 5 ROWS:")
print(df.tail())

print("\nDATASET INFORMATION:")
df.info()

print("\nSTATISTICAL DESCRIPTION:")
print(df.describe())

#Check missing values
print("\nMISSING VALUES:")
print(df.isnull().sum())

#Create Age column
df['Age'] = 2025 - df['Year_Birth']

#Bar plot 
plt.figure()
df['Age'].value_counts().head(10).plot(kind='bar')
plt.title("Age Distribution (Top 10)")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

#Box plot
plt.figure()
df['Income'].plot(kind='box')
plt.title("Income Distribution")
plt.ylabel("Income")
plt.show()

#Calculate Total Spending
df['Total_Spending'] = (
    df['MntWines'] +
    df['MntFruits'] +
    df['MntMeatProducts'] +
    df['MntFishProducts'] +
    df['MntSweetProducts'] +
    df['MntGoldProds']
)

#Box plot 
plt.figure()
df['Total_Spending'].plot(kind='box')
plt.title("Customer Spending Pattern")
plt.ylabel("Total Spending")
plt.show()


```
