*DESCRIPTIVE ANALYSIS:*

1. **Histogram  Happiness Score**

The histogram shows the distribution of happiness scores across countries. 2. **Bar Plot**

The bar chart clearly compares the top 10 happiest countries. 3. **Box Plot  Happiness Score**

The box plot represents the spread and variability of happiness scores. The median lies closer to the upper quartile, indicating slight skewness. A few outliers are present, showing countries with exceptionally high or low happiness compared to the rest.

4. **Scatter Plot  GDP per Capita vs Happiness Score**

5. **Line Plot  Happiness Rank vs Happiness Score**

The line plot shows an inverse relationship between rank and happiness score. As the happiness rank increases (lower rank position), the happiness score decreases steadily. This confirms the ranking logic used in the dataset.

6. **Pie Chart**

The pie chart shows the proportion of happiness scores among the top 5 happiest countries. The percentages are relatively close, indicating that these top countries have similar happiness levels, with no single country dominating completely.

7. **Heatmap  Correlation Analysis**

The heatmap visualizes correlations among numerical variables. Strong positive correlations are observed between happiness score and factors such as GDP per capita, health, and social support. This indicates that economic stability and well-being play a major role in determining happiness.

**Overall Interpretation**

Based on the visualizations, happiness levels vary significantly across countries and are influenced by multiple factors rather than a single variable. Economic strength, health, and social support collectively contribute to higher happiness scores.
