```python
print("\nK.Nivedha(24BAD082)")
print("\nEXPERIMENT 1")
print("\nSCENARIO 1: E-COMMERCE SALES DATA\n")

import pandas as pd
import matplotlib.pyplot as plt

#Load the dataset
df = pd.read_csv(r"C:\Users\nived\OneDrive\Desktop\dataset python\data.csv", encoding="ISO-8859-1")

#first 5 rows
print("FIRST 5 ROWS:")
print(df.head())

#last 5 rows
print("\nLAST 5 ROWS:")
print(df.tail())

#Dataset information
print("\nDATASET INFO:")
df.info()

print("\nSTATISTICAL SUMMARY:")
print(df.describe())

print("\nMISSING VALUES IN EACH COLUMN:")
print(df.isnull().sum())

product_sales = df.groupby("Description")["Quantity"].sum().head(10)

#Bar chart
plt.figure()
product_sales.plot(kind='bar')
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.title("Top Selling Products")
plt.show()

#Line chart
plt.figure()
product_sales.plot(kind='line')
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.title("Sales Trend of Products")
plt.show()
```
