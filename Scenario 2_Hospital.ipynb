```python
print("\nK.Nivedha(24BAD082)")
print("\nEXPERIMENT 1")
print("\nSCENARIO 2: HOSPITAL PATIENT RECORDS\n")

import pandas as pd
import matplotlib.pyplot as plt

#Load the dataset
df2 = pd.read_csv(r"C:\Users\nived\Downloads\diabetes.csv")

#first 5 rows
print("FIRST 5 ROWS:")
print(df2.head())

#Dataset information
print("\nDATASET INFO:")
df2.info()

#Check missing values
print("\nMISSING VALUES IN EACH COLUMN:")
print(df2.isnull().sum())

#Statistical summary
print("\nSTATISTICAL SUMMARY:")
print(df2.describe())

#Histogram 
plt.figure()
plt.hist(df2["Glucose"], bins=20)
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.title("Glucose Level Distribution")
plt.show()

plt.figure()
plt.hist(df2["Age"], bins=10)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution")
plt.show()

#Boxplot 
plt.figure()
plt.boxplot(df2["Glucose"])
plt.title("Boxplot of Glucose Levels")
plt.show()

plt.figure()
plt.boxplot(df2["Age"])
plt.title("Boxplot of Age")
plt.show()

```
