```python
print("\nK.Nivedha(24BAD082)")
print("\nEXPERIMENT 1")
print("\nSCENARIO 3: HOUSING DATASET\n")

import pandas as pd
import matplotlib.pyplot as plt

#Load the dataset
df3 = pd.read_csv(r"C:\Users\nived\Downloads\Housing (1).csv")

#first 5 rows
print("FIRST 5 ROWS:")
print(df3.head())

#dataset information
print("\nDATASET INFO:")
df3.info()

#Check missing values
print("\nMISSING VALUES IN EACH COLUMN:")
print(df3.isnull().sum())

# Statistical summary
print("\nSTATISTICAL SUMMARY:")
print(df3.describe())

#Scatter plot
plt.figure()
plt.scatter(df3["area"], df3["price"])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()

#Correlation matrix
numeric_df = df3.select_dtypes(include=['int64', 'float64'])

#Heatmap for correlation
corr = numeric_df.corr()

print("\nCORRELATION MATRIX:")
print(corr)


```
